```python

def adjust_rating(rating):
    if rating < 1:
        return 1
    elif rating > 5:
        return 5
    else:
        return rating

# Apply the function to the 'rating' column
rat["rating"] = rat["rating"].apply(adjust_rating)

```

```python
allowed_devices = ['Mobile', 'Laptop', 'Smart TV']

# Filter rows where the device is one of the allowed ones
watch = watch[watch['device_type'].isin(allowed_devices)]


cleaned_file_path6 = r"C:\Users\ramch\OneDrive\Documents\clean_csv_project\cleaner_watch.csv"

watch.to_csv(cleaned_file_path6, index=False)


```

```python

def adjust_negative_values(row):
    if row['amount'] < 0:
        
        if row['plan_type'] == 'Student Plan':
            return 4.9900  
        elif row['plan_type'] == 'Standard':
            return 7.9900  
        elif row['plan_type'] == 'Premium':
            return 12.9900 
        else:
            return 0 
    return row['amount']
    
sub['amount'] = sub.apply(adjust_negative_values, axis=1)


cleaned_file_path8 = r"C:\Users\ramch\OneDrive\Documents\clean_csv_project\cleaner_sub.csv"
sub.to_csv(cleaned_file_path8, index=False)
```

```python
average_age = round(users['age'].mean())


def fix_age(row):
    if row['age'] < 1 or row['age'] > 100:
        return average_age 
    return row['age']  
    
users['age'] = users.apply(fix_age, axis=1)
```
